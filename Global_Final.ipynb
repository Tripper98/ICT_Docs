```python
# Challenge 1 
# import tensorflow as tff 
import tensorflow.compat.v1  as tf

width = 56 
height = 64
channel = 3

# inference
tf.disable_eager_execution()

X = tf.placeholder(tf.float32, shape=[None,width,
                                    height,
                                    channel], name ='inputs')
y_= tf.placeholder(tf.float32, shape=[None,], name ='labels')

# convolutional layer 1 

with tf.variable_scope('convolutional-1',reuse=tf.AUTO_REUSE) : 
    conv1_weights = tf.get_variable('weight'
                                    ,[5,5,3,20],
                                   initializer = tf.truncated_normal_initializer(stddev=0.1))
    conv1_biases = tf.get_variable('bias',
                                  [20],
                                  initializer = tf.constant_initializer(0.0))
    
    conv1 = tf.nn.conv2d(X,conv1_weights, 
                         strides=[1,1,1,1],
                         padding = 'VALID')
    relu1 = tf.nn.relu(tf.nn.bias_add(conv1,conv1_biases))

with tf.name_scope('max-pooling-1') : 
    pool1 = tf.nn.max_pool(relu1, 
                          ksize = [1,2,2,1], 
                          strides= [1,2,2,1],
                          padding = 'VALID')
# Challenge 1 
# Convolutional layer 2 
with tf.variable_scope('convolutional-2',reuse=tf.AUTO_REUSE) : 
    conv2_weights = tf.get_variable('weight',
                                    [3,3,20,30],
                                   initializer = tf.truncated_normal_initializer(stddev=0.1))
    conv2_biases = tf.get_variable('bias',
                                  [30],
                                  initializer = tf.constant_initializer(0.0))
    
    conv2 = tf.nn.conv2d(pool1,conv2_weights, 
                         strides=[1,1,1,1],
                         padding = 'VALID')
    relu2 = tf.nn.relu(tf.nn.bias_add(conv2,conv2_biases))
    
# Challenge 2 
with tf.name_scope('max-pooling-2') : 
    pool2 = tf.nn.max_pool(relu2, 
                          ksize = [1,2,2,1], 
                          strides= [1,2,2,1],
                          padding = 'VALID')

# Convolutional layer 3 
with tf.variable_scope('convolutional-3',reuse=tf.AUTO_REUSE) : 
    conv3_weights = tf.get_variable('weight',
                                    [3,3,30,40],
                                   initializer = tf.truncated_normal_initializer(stddev=0.1))
    conv3_biases = tf.get_variable('bias',
                                  [40],
                                  initializer = tf.constant_initializer(0.0))
    
    conv3 = tf.nn.conv2d(pool2,conv3_weights, 
                         strides=[1,1,1,1],
                         padding = 'VALID')
    relu3 = tf.nn.relu(tf.nn.bias_add(conv3,conv3_biases))

# Convolutional layer 4 
with tf.variable_scope('convolutional-4',reuse=tf.AUTO_REUSE) : 
    conv4_weights = tf.get_variable('weight',
                                    [3,3,40,50],
                                   initializer = tf.truncated_normal_initializer(stddev=0.1))
    conv4_biases = tf.get_variable('bias',
                                  [50],
                                  initializer = tf.constant_initializer(0.0))
    
    conv4 = tf.nn.conv2d(relu3,conv4_weights, 
                         strides=[1,1,1,1],
                         padding = 'VALID')
    relu4 = tf.nn.relu(tf.nn.bias_add(conv4,conv4_biases))
    
# Challenge 2 
with tf.name_scope('max-pooling-3') : 
    pool3 = tf.nn.max_pool(relu4, 
                          ksize = [1,2,2,1], 
                          strides= [1,2,2,1],
                          padding = 'VALID')
    
# Challenge 3 

featuresize = 5*6*50 
deepfeatures = tf.reshape(pool3,[-1,featuresize],name='features')

# ADDED
regularizer = tf.keras.regularizers.L2(0.0001)
dropout = False 

# Coding a fully connected Layer .
with tf.variable_scope('fully-connected-1',reuse=tf.AUTO_REUSE) :
    full1_weights = tf.get_variable('weight',
                                    [featuresize,200],
                                   initializer = tf.truncated_normal_initializer(stddev=0.1))
    full1_biases = tf.get_variable('bias',
                                  [200],
                                  initializer = tf.constant_initializer(0.0))
    full1 = tf.nn.relu(tf.matmul(deepfeatures,full1_weights)+full1_biases)
    
    if dropout : 
        full1 = tf.nn.dropout(full1,0.5)
    if regularizer != None :
        tf.add_to_collection('losses',regularizer(full1_weights))
    
    
with tf.variable_scope('fully-connected-2',reuse=tf.AUTO_REUSE) :
    full2_weights = tf.get_variable('weight',
                                    [200,200],
                                   initializer = tf.truncated_normal_initializer(stddev=0.1))
    full2_biases = tf.get_variable('bias',
                                  [200],
                                  initializer = tf.constant_initializer(0.0))
    full2 = tf.nn.relu(tf.matmul(full1,full2_weights)+full2_biases)
    
    if dropout : 
        full2 = tf.nn.dropout(full2,0.5)
    if regularizer != None :
        tf.add_to_collection('losses',regularizer(full2_weights))
    
    
with tf.variable_scope('fully-connected-3',reuse=tf.AUTO_REUSE) :
    full3_weights = tf.get_variable('weight',
                                    [200,200],
                                   initializer = tf.truncated_normal_initializer(stddev=0.1))
    full3_biases = tf.get_variable('bias',
                                  [200],
                                  initializer = tf.constant_initializer(0.0))
    logit = tf.matmul(full2,full3_weights)+full3_biases
    
    if regularizer != None :
        tf.add_to_collection('losses',regularizer(full3_weights))
    
    
# Softmax 
softmax_result = tf.nn.softmax(logit,name='outputs')
```

```python
with tf.Session() as sess : 
    writer = tf.summary.FileWriter('./models/tbdir1/',sess.graph)
```

```python
# loss function 
loss = tf.nn.sparse_softmax_cross_entropy_with_logits(logits = logit, labels = tf.dtypes.cast(y_,tf.int32))

# loss regularizer 
loss_reg = tf.add_n(tf.get_collection('losses'))
loss = loss + loss_reg

# optimizer 
train_op = tf.train.AdamOptimizer(learning_rate = 0.001).minimize(loss)
```
